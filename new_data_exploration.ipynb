# Quick Data Exploratory: New Dataset

Dataset built by mishra, sandhya; Soni, Devpriya (2022), “SMS PHISHING DATASET FOR MACHINE LEARNING AND PATTERN RECOGNITION”, Mendeley Data, V1, doi: 10.17632/f45bkkt8pr.1

Available in: https://data.mendeley.com/datasets/f45bkkt8pr/1

```python
import pandas as pd

# Load the dataset
data_path = './data/Dataset_5971.csv'
df = pd.read_csv(data_path)

# Display the first few rows of the DataFrame to get an initial idea of the data
print(df.head())
```

```python
# General overview of the data
print(df.info())
```

```python
# Descriptive statistics for numerical columns
print(df.describe())
```

### 1. Target Variable Analysis

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Count each category in the target variable
label_counts = df['LABEL'].value_counts()

# Bar plot for the distribution of LABEL
plt.figure(figsize=(10, 6))
sns.barplot(x=label_counts.index, y=label_counts.values, palette='viridis')
plt.title('Distribution of Target Variable LABEL')
plt.xlabel('LABEL Categories')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.show()
```

## 2. Categorical Variables Analysis

```python
# Bar plots for categorical variables
categorical_vars = ['URL', 'EMAIL', 'PHONE']

for var in categorical_vars:
    counts = df[var].value_counts()

    plt.figure(figsize=(8, 4))
    sns.barplot(x=counts.index, y=counts.values, palette='viridis')
    plt.title(f'Distribution of {var} Variable')
    plt.xlabel(var + ' Categories')
    plt.ylabel('Frequency')
    plt.xticks(rotation=45)
    plt.show()
```

## 3. Text Analysis: Wordcloud

```python
from wordcloud import WordCloud

# Combine all text into one large string and create a word cloud
text = ' '.join(df['TEXT'].dropna())
wordcloud = WordCloud(width=800, height=400, background_color ='white').generate(text)

# Display the word cloud
plt.figure(figsize=(10, 7))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Word Cloud for TEXT')
plt.show()
```
