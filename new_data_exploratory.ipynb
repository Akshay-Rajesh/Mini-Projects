# Quick Exploratory over new dataset

```python
import pandas as pd

# Load dataset from "data" folder
data_path = './data/Dataset_5971.csv'
df = pd.read_csv(data_path)

# Show a sample
print(df.head())
```

```python
# Data General Description
print(df.info())
```

```python
# Descriptive Statistics
print(df.describe())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Counting each label for LABEL
label_counts = df['LABEL'].value_counts()

# LABEL categories barplot distribution
plt.figure(figsize=(10, 6))
sns.barplot(x=label_counts.index, y=label_counts.values, palette='viridis')
plt.title('Target Variable distribution')
plt.xlabel('Classes of LABEL')
plt.ylabel('Freq')
plt.xticks(rotation=45)
plt.show()
```

```python
# Categorical Var. Bars
categorical_vars = ['URL', 'EMAIL', 'PHONE']

for var in categorical_vars:
    counts = df[var].value_counts()

    plt.figure(figsize=(8, 4))
    sns.barplot(x=counts.index, y=counts.values, palette='viridis')
    plt.title(f'Distribution of {var}')
    plt.xlabel('Categories of ' + var)
    plt.ylabel('Freq')
    plt.xticks(rotation=45)
    plt.show()
```

```python
from wordcloud import WordCloud

# It combines all the text in one only before to create the word cloud
text = ' '.join(df['TEXT'].dropna())
wordcloud = WordCloud(width=800, height=400, background_color ='white').generate(text)

# Wordcloud visuals
plt.figure(figsize=(10, 7))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Wordcloud of New Dataset')
plt.show()
```
